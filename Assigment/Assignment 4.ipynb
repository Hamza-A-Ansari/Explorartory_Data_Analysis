```python
import numpy as np
import pandas as pd
```

```python
actor=pd.read_csv('./Datasets/actor.csv')
actor
```

```python
film_actor=pd.read_csv('./Datasets/film_actor.csv')
film_actor
```

## Left Join:


```python
actor.merge(film_actor, on='actor_id', how='left')
```

## Right Join

```python
actor.merge(film_actor, on='actor_id', how='right')
```

## Inner Join

```python
actor.merge(film_actor, on='actor_id')
```

## Outer Join

```python
actor.merge(film_actor, on='actor_id',how="outer")
```
